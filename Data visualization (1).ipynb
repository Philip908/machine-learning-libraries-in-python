```python
x = np.array([1000,2000,3000,4000])
y = np.array([10,20,30,40])
plt.plot(x,y,"o",color ="red",mec="blue",ms=15)
plt.xlabel("area",fontsize=20)
plt.ylabel("price",fontsize=20)
plt.title("Relation B/W price and area",fontsize=20)
plt.grid(axis = "y", color = "k",linestyle = "--")
plt.grid(axis = "x", color = "k",linestyle = "--")
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([10,20,30,40,50])
plt.bar(x,y,color="hotpink")
plt.xlabel("Sales")
plt.ylabel("Income")
plt.title("Company info")
plt.show()
```
